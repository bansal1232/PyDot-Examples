# Directory structure plotter

## Installation

- pydot3
- graphviz

```python
import pydot
import os
```

```python
rootDir = "/home/nikhil/Desktop/python-resources"
```

```python
G = pydot.Dot(graph_type="digraph")
```

```python
node = pydot.Node(rootDir.split('/')[-1], style="filled", fillcolor="green")
G.add_node(node)
```

```python
for root, dirs, files in os.walk(rootDir):
    
    for subdir in dirs:
        node = pydot.Node(subdir, style="filled", fillcolor="yellow")
        G.add_node(node)
        edge = pydot.Edge(root.split('/')[-1], subdir)
        G.add_edge(edge)
        
    for file in files:
        node = pydot.Node(file, style="filled", fillcolor="orange")
        G.add_node(node)
        edge = pydot.Edge(root.split('/')[-1], file)
        G.add_edge(edge)
```

```python
from IPython.display import Image, display

im = Image(G.create_png())
display(im)
```
